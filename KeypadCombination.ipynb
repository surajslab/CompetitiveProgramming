# Keypad Combination Problem
## Based on Recursion

```python


def get_string(smallno):
    
    if smallno==2:
        return "abc"
    elif smallno==3:
        return "def"
    elif smallno==4:
        return "ghi"
    elif smallno==5:
        return "jkl"
    elif smallno==6:
        return "mno"
    elif smallno==7:
        return "pqrs"
    elif smallno==8:
        return "tuv"
    elif smallno==9:
        return "wxyz"
    else:
        return " "

        





def keypad(number):
    
    if number==0 or number==1:
        output=[]
        output.append("")
        return output
    
    smallno=number%10
    bigno=number//10
    output=[]
    bigoutput=keypad(bigno)
    
    sstring=get_string(smallno)
    
    
    for i in sstring:
        for j in bigoutput:
            output.append(j+i)
    
    return output
    
    

inputs=int(input())

ans=keypad(inputs)
for i in ans:
    print(i)

    

  
```
